```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('sonar_dataset.csv', header=None)
df.sample(5)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df[60].value_counts()
```

```python
X = df.drop(60, axis='columns')
y = df[60]
```

```python
X
```

```python
y
```

```python
y = pd.get_dummies(y, drop_first=True)
y   # R --> 1, M --> 0
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.shape, X_test.shape
```

## Build the ANN

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# Round the values to nearest integer, 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
print(y_test[:10])
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

## Model with dropout layers

```python
model_dropout = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.7),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.7),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model_dropout.summary()
```

```python
model_dropout.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model_dropout.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model_dropout.evaluate(X_test, y_test)
```

```python
y_pred = model_dropout.predict(X_test).reshape(-1)
print(y_pred[:10])

# Round the values to nearest integer, 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
print(y_test[:10])
```

```python
print(classification_report(y_test, y_pred))
```
